```python
import matplotlib.pyplot as plt
import math

# grab one batch
specs, labels = next(iter(train_loader))  # specs: [B,1,H,W], labels: [B]

B, _, H, W = specs.shape
cols = 4
rows = math.ceil(B / cols)

fig, axes = plt.subplots(rows, cols, figsize=(cols * 3, rows * 2.5))
axes = axes.flatten()

for i in range(B):
    ax = axes[i]
    spec_i = specs[i].squeeze(0).cpu().numpy()  # [H, W]
    ax.imshow(spec_i, aspect="auto", origin="lower")
    ax.set_title(f"Label {labels[i].item()}")
    ax.set_xticks([])
    ax.set_yticks([])

# hide any leftover axes
for j in range(B, len(axes)):
    axes[j].axis("off")

plt.tight_layout()
plt.show()
```
